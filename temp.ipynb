```python
d1=pd.read_csv(r"C:\Users\tgup1\Downloads\dwh.csv")
def pdate(x):
    if len(x.split("-")[0])==4:
        return datetime.strptime(x,"%Y-%m-%d").weekday()
    else:
        return datetime.strptime(x,"%m-%d-%Y").weekday()
import calendar
dic ={}
for x in range(7):
    dic[x]=calendar.day_name[x]
print(dic)

d1["weekday"]=d1["ActivationDate"].apply(lambda x: dic[pdate(x)])
```
