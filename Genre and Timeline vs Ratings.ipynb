```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ast import literal_eval
from datetime import datetime
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
metadata = pd.read_csv("dataset/movies_metadata.csv",low_memory=False)
```

# Preprocessing

```python
#convert genres to readable format as list of genres
metadata["genres"] = metadata["genres"].apply(literal_eval).apply(lambda x: [i["name"] for i in x] 
                                                                  if isinstance(x,list) else [])
```

```python
# metadata["release_date"].apply(lambda x: datetime.strptime(x,"%d-%m-%Y") if x!="NaN" else "")
```

```python
# function to convert string date to datetime.
def time_format_check(x):
    try:
        datetime.strptime(x,"%d-%m-%Y")
        return True
    except:
        try:
            datetime.strptime(x,"%Y-%m-%d")
            return True
        except:
            return False
df=metadata[metadata["release_date"].apply(time_format_check)][["release_date","genres","vote_average"]]\
    .dropna(subset=["vote_average"])[metadata["genres"].str.len()!=0]
```

# Year vs Month releases

```python
def month_quater(x):
    if 0<= x and x < 4 :
        return 'q1'
    if 4<= x and x < 6 :
        return 'q2'
    if 6<= x and x < 10 :
        return 'q3'
    if 10<= x and x <= 12:
        return 'q4'
df["quater"]=df["month"].apply(month_quater)
# df[["year","quater","month"]].sort_values(by=["year","quater"])
# df.groupby(["year","quater"])["month"].count().to_csv("year_quater.csv")
# df.groupby(["year","quater"])["vote_average"].mean().to_csv("year_quater_avg.csv")
```

```python
# df.groupby(["year"])["month"].count().to_csv("year_movie_count.csv")
for genre in genres:
    x=df.groupby(["year","month"])["vote_average",genre].mean()
    x=x[x[genre]!=0]
    x.to_csv("genre2/"+genre+".csv")
# df.groupby(["year","month"]).mean().to_csv("genre/all.csv")
```

```python
year_range = [x for x in range(1874,2025,10)]
df["year_range"]=pd.cut(df.year,bins=year_range,right=False)
month_range = [x for x in range(1,14,3)]
df["month_range"]=pd.cut(df.month,bins=month_range,right=False)
```
