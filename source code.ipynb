# Genre Word Cloud

```python

genre_dict={}
count_genre = {}
# drop rows where genre is na
metadata.dropna(subset=['genres'],inplace=True)

for x in metadata["genres"].as_matrix():
    x = x.replace("'","\"")
    y = json.loads(x)
    for temp in y:
        genre_dict[temp['id']] = temp['name']
        if temp['name'] in count_genre.keys():
            count_genre[temp['name']] +=1
        else:
            count_genre[temp['name']] =1
        
                   
print(genre_dict)
print(count_genre)

# create word cloud
wordcloud = WordCloud(height=400,width=500).fit_words(count_genre).to_file("genre_wordCloud.jpg")
```

```python
#Languages used in movies
data = metadata["spoken_languages"]
languages = {}
count_languages={}
for x in data:
    try:
        x = x.replace("'","\"")
        y = json.loads(x)
        for temp in y:
            languages[temp["iso_639_1"]]=temp["name"]
            if temp["iso_639_1"] in count_languages.keys():
                count_languages[temp["iso_639_1"]] +=1
            else:
                count_languages[temp["iso_639_1"]] = 1
    except:
        pass
print(languages)


       
        
   
```

```python
#Plotting Graph for languages used
import operator
sorted_dic = sorted(count_languages.items(),key=operator.itemgetter(1),reverse=True)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(count_languages.keys(),count_languages.values(),color='g')
#todo need to corret the graph axis
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Movie profits 
df = metadata[["title","revenue","budget","popularity"]]
df = df.dropna(subset=['revenue',"budget"])
df["profit"] = df["revenue"].astype('int') - df["budget"].astype('int')
x=df.sort_values(by=["profit"],axis=0,ascending=False).iloc[:10,:]
pprint(x)




```

```python
# merging credits and metadata csv
def remover(x):
    if "-" in x:
        return 
    else:
        return int(x)
metadata["id"]=metadata["id"].apply(remover)
metadata=metadata.dropna(subset=["id"])

metadata.sort_values(by=["id"],axis=0,inplace=True)

df_credits = pd.read_csv("dataset/credits.csv",low_memory=False)
df_credits.sort_values(by=["id"],axis=0,inplace=True)

# df = pd.merge(left=metadata,right=df_credits,left_on="id",right_on="id")
df = metadata.merge(df_credits,on="id",how="inner")
df['castname'] = df['cast'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list)
else [])
```

```python
df.columns
```

```python
actor = "Scarlett Johansson"
best_cast = df[["revenue","budget","castname","title"]]
best_cast["budget"]=best_cast["budget"].astype("int64")
best_cast.dropna(subset=["revenue","budget"],inplace=True)
best_cast["profit"] = best_cast["revenue"] - best_cast["budget"]
best_cast=best_cast[(best_cast["revenue"]!=0)&(best_cast["budget"]!=0)]
best_cast["names"]=best_cast["castname"].apply(lambda x: ",".join(x))
performer = best_cast[best_cast["names"].str.contains(actor)]
performer.sort_values(by="profit",ascending=False)
        
```
